```python
import meep as mp
```

```python
cell = mp.Vector3(4, 1, 0)
```

```python
geometry = []
```

```python
vol = mp.Volume(mp.Vector3(-1.0,0.0,0.0), mp.Vector3(0.01,1.0,0.0))
sources = [
    mp.Source(
        mp.ContinuousSource(frequency=2.0), component=mp.Ez, volume=vol
    )
]
```

```python
pml_layers = [mp.PML(0.1)]
```

```python
resolution = 100
```

```python
sim = mp.Simulation(
    cell_size=cell,
    boundary_layers=pml_layers,
    geometry=geometry,
    sources=sources,
    resolution=resolution,
)
```

```python
from matplotlib import pyplot as plt

%matplotlib inline
plt.figure(dpi=100)
sim.plot2D()
plt.show()
```

```python
sim.run(until=2)
```

It should finish in less than a second. We can analyze and visualize the fields using `plot2D`:

```python
plt.figure(dpi=100)
sim.plot2D(fields=mp.Ez)
plt.show()
```
